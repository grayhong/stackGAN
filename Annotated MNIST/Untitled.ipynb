```python
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np

from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data', one_hot=False)

%matplotlib inline
```

```python
def get_num(num, mnist, train=False):
    if train:
        num_idx = (mnist.train.labels == num)
        return mnist.train.images[num_idx]
    else:
        num_idx = (mnist.test.labels == num)
        return mnist.test.images[num_idx]
```

```python
ones = get_num(7, mnist, False)

plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.imshow(ones[i].reshape([28, 28]), interpolation='NONE', vmin=0, vmax=1, cmap=plt.get_cmap('gray'))
plt.tight_layout()
```
